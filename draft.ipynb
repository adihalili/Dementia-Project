```python
dmn_df['Education_Level']= dmn_df['Education_Level'].replace(['No School', 'Primary School','Secondary School', 'Diploma/Degree'], [0,1,2,3])
dmn_df['Sleep_Quality']= dmn_df['Sleep_Quality'].replace(['Poor','Good'], [0,1])
dmn_df['Smoking_Status']= dmn_df['Sleep_Quality'].replace(['Current Smoker', 'Former Smoker', 'Never Smoked'], [2, 1, 0])
```

```python
corrs= dmn_df.corr(method= 'kendall', numeric_only= True)
plt.figure(figsize=(20,10))
sns.heatmap(corrs, cmap= 'coolwarm', center= 0, annot = True)
```

```python
smoking_sleep= ['Smoking_Status', 'Education_Level', 'Cognitive_Test_Scores', 'Sleep_Quality']
corr = dmn_df[smoking_sleep].corr(method= 'kendall', numeric_only= True)
fig, ax = plt.subplots(figsize=(20,5)) 
sns.heatmap(corr,annot=True, cmap = 'Blues', linewidth = 0.6)
plt.title('Correlation matrix')
plt.show()
```
